## 애플 cpu
- cpu 이름 애플로 채우기
- 제품이름에서 추출한 칩셋 등급을 챗봇의 도움을 받아 라이젠과 인텔의 cpu와 함께 3,5,7,9 등급을 맞춰 카테고리화

```python
import re

def extract_apple_cpu(name):
    match = re.search(r'(M\d)(프로|맥스|울트라)?', name)
    if match:
        chip = match.group(1)           # M2, M3, M4, M5
        variant = match.group(2) or ''  # 프로, 맥스, 울트라, 없으면 빈 문자열
        return chip + variant
    return None

apple_mask = 다나와2['브랜드'] == 'APPLE'

# CPU브랜드 채우기
다나와2.loc[apple_mask, 'CPU브랜드'] = 'APPLE'

# CPU등급 채우기
다나와2.loc[apple_mask, 'CPU등급'] = 다나와2.loc[apple_mask, '제품명'].apply(extract_apple_cpu)
```
